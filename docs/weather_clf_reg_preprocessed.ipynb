```python
# Import necessary libraries for data manipulation, machine learning models, and metrics
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Load the preprocessed weather dataset
nome_file = "./weather_clf_reg_preprocessed.csv"
df = pd.read_csv(nome_file)

df
```

Preparazione dati

```python
# Separate features (X) and target variable (y)
X = df.drop(columns=["RainfallTomorrow"])
y = df["RainfallTomorrow"]
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train
```

```python
# Standardize the numerical features by removing the mean and scaling to unit variance
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

Regressore Unico

```python
# Perform Grid Search Cross-Validation to find the best hyperparameters for the general Random Forest Regressor
search_reg = RandomForestRegressor()

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, None],
}

grid_search = GridSearchCV(search_reg, param_grid, cv=5, n_jobs = -1)
grid_search.fit(X_train, y_train)

print("Risultati medi per tutte le combinazioni:", grid_search.cv_results_["mean_test_score"])
print("Migliori parametri trovati:", grid_search.best_params_)
print("Punteggio migliore:", grid_search.best_score_)
```

```python
# Train a Random Forest Regressor using a specific set of parameters
reg1 = RandomForestRegressor(max_depth = None, n_estimators = 200, n_jobs= -1)
reg1.fit(X_train, y_train)
y_pred1 = reg1.predict(X_test)
```

```python
# Evaluate the regressor performance (R^2 score and MAE)
print(reg1.score(X_test, y_test))
print(mean_absolute_error(y_pred1, y_test))
```

Classificatore + Regressore

```python
# Dati per allenare il classificatore
X_clf = X_train
y_clf = (y_train > 0)
```

```python
# Dati per allenare il regressore
X_reg = X_train[y_clf]
y_reg = y_train[y_clf]
```

Classificatore

```python
# Perform Grid Search to find best parameters for the Classifier
search_clf = RandomForestClassifier()

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, None],
}

grid_search = GridSearchCV(search_clf, param_grid, cv=5, n_jobs = -1)
grid_search.fit(X_clf, y_clf)

print("Risultati medi per tutte le combinazioni:", grid_search.cv_results_["mean_test_score"])
print("Migliori parametri trovati:", grid_search.best_params_)
print("Punteggio migliore:", grid_search.best_score_)
```

Regressore

```python
# Perform Grid Search to find best parameters for the Regressor (on filtered data)
search_reg = RandomForestRegressor()

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, None],
}

grid_search = GridSearchCV(search_reg, param_grid, cv=5, n_jobs = -1)
grid_search.fit(X_reg, y_reg)

print("Risultati medi per tutte le combinazioni:", grid_search.cv_results_["mean_test_score"])
print("Migliori parametri trovati:", grid_search.best_params_)
print("Punteggio migliore:", grid_search.best_score_)
```

Allenamento classificatore e regressore con i migliori parametri trovati

```python
# Train the Classifier with best parameters found
clf = RandomForestClassifier(max_depth = None, n_estimators = 200, n_jobs = -1)
reg = RandomForestRegressor(max_depth = None, n_estimators = 200, n_jobs= -1)
clf.fit(X_clf, y_clf)
reg.fit(X_reg, y_reg)
```

Testing

```python
# Predict if it will rain, then predict the rain amount only for predicted rainy days
y_pred_clf = clf.predict(X_test)
y_pred_reg = reg.predict(X_test)
y_pred2 = [0 if c==0 else r for (c, r) in zip(y_pred_clf, y_pred_reg)]
```

```python
print(mean_absolute_error(y_pred2, y_test))
```

Confronto risultati

```python
# MAE
print(mean_absolute_error(y_pred1, y_test), mean_absolute_error(y_pred2, y_test))

# MSE
print(mean_squared_error(y_pred1, y_test), mean_squared_error(y_pred2, y_test))

# Accuracy
y_pred1_int = [int(e) for e in y_pred1]
y_pred2_int = [int(e) for e in y_pred2]
print(sum(y_pred1_int==y_test)/len(y_test), sum(y_pred2_int==y_test)/len(y_test))
```
